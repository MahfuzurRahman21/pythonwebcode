```python
trans_cost=[150,150,150,150,150]
space_cost=[50,50,50,50,50]
staff_cost=[60,60,60,60,60]
mat_cost=[5.4,4.5,6.6,4.95,5.1]

total_inventory_cost = [sum(i) for i in zip(trans_cost, space_cost,staff_cost,cake_price,mat_cost )]  
print(total_inventory_cost)
res = "\n".join("the {} item total production price is {}".format(x, y) for x, y in zip(cake_items, total_inventory_cost))
print(res)  

```

```python
selling_price=[780,600,800,650,600]
discounted=[39.0,30.0,40.0,45.5,42.0]
aft_dis_price=[]
for i,j in zip(selling_price,discounted):
  aft_dis_price.append(i-j)
print(aft_dis_price)  
print("\n".join("the {} item after discounted price is {}".format(x,y) for x,y in zip(cake_items,aft_dis_price))) 
```

```python
print(type(aft_dis_price))
```

```python
#total selling price after discount
sold=[5,7,10,5,9]
total_sel_Pr=[]
for i,j in zip(aft_dis_price,sold):
  total_sel_Pr.append(i*j)
print(total_sel_Pr)  
print("\n".join("the {} item total sold price is {}".format(x,y) for x,y in zip(cake_items,total_sel_Pr))) 
```

```python
#amount of profit/loss in taka
selling_price=[780,600,800,650,600]
total_inventory_cost=[445.4, 414.5, 486.6, 429.95, 435.1]
profit=[]
for i,j in zip(selling_price,total_inventory_cost):
  profit.append(i-j)
print(profit)  
print("\n".join("the {} item amount of profit is {} BDT".format(x,y) for x,y in zip(cake_items,profit))) 
```

```python
#profit or loss in percentage
selling_price=[780,600,800,650,600]
total_inventory_cost=[445.4, 414.5, 486.6, 429.95, 435.1]
profit=[]
for i,j in zip(selling_price,total_inventory_cost):
  profit.append(i*100/j)
print(profit)  
print("\n".join("the {} item amount of profit is {:.2f} %".format(x,y) for x,y in zip(cake_items,profit))) 
```
